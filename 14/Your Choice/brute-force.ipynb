```python
# neccessary definitions for graph node and edge
class Graph:
    def __init__(self,n, nodes):
        self.nodes = []
        self.edges = []
        self.n = n
        for i in range(n):
            self.nodes.append(nodes[i])
    def add_edge(self,edge):
        self.edges.append(edge)
    def get_node(self,index):
        return self.nodes[index]
    def get_edge(self,index):
        return self.edges[index]
    def get_nodes(self):
        return self.nodes
    def get_edges(self):
        return self.edges
    def get_n(self):
        return self.n
class Node:
    def __init__(self,index):
        self.index = index
        self.edges = []
    def add_edge(self,edge):
        self.edges.append(edge)
    def get_index(self):
        return self.index
    def get_edges(self):
        return self.edges
    def get_edge(self,index):
        return self.edges[index]
class Edge:
    def __init__(self,node1,node2):
        self.node1 = node1
        self.node2 = node2
    def get_node1(self):
        return self.node1
    def get_node2(self):
        return self.node2
```

```python
class Problem:
    def __init__(self,graph,end):
        self.graph = graph
        self.end = end
    def get_graph(self):
        return self.graph
    def get_end(self):
        return self.end
```

```python
def generate_problem(lines, start, input_number,max_box):
    # generate problem
    # lines: the lines of file
    # start: the start index of problem
    # input_number: the number of input
    # max_box: the max number of box
    # return: the problem
    #create a graph with input number of nodes
    
    #create a list of nodes
    nodes = []
    for i in range(input_number):
        if (i == max_box):
            max_node = Node(i)
            nodes.append(max_node)
        else:
            nodes.append(Node(i))
    edges = []
    for i in range(input_number):
        for j in range(input_number):
            if(lines[i+start][j] == 'Y'):
                #print("edge",i,j)
                edge = Edge(nodes[i],nodes[j])
                edges.append(edge)
                nodes[i].add_edge(edge)
    graph = Graph(input_number, nodes)
    #create a list of edges
    #add edges to graph
    for edge in edges:
        graph.add_edge(edge)
    # add  nodes to graph

    return graph, max_node
```

```python
with open('small-testdata.in') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
lines = lines[1:]
file_length = len(lines)
i = 0
problems = []
while i < file_length:
    line = lines[i].split(" ")
    #print(i)
    input_number = line[0]
    #print(input_number)
    max_box = int(line[1])
    #print generated problem
    graph, max_node = generate_problem(lines, i+1, int(input_number), int(max_box))
    problem = Problem(graph, max_node)
    problems.append(problem)
    i+= int(input_number)+1
```

```python
def recursive_search_on_graph(node, max_node, visited_node):
    visited_node.append(node)
    #print(visited_index)
    #if we arrive the maximum node and there are no where else to go that it is a success
    if node.get_index() == max_node.get_index():
        print("arrived at max node")
        for edge in node.get_edges():
            if edge.get_node1().get_index() == node.get_index():
                if(edge.get_node2() not in visited_node):
                    #check if it way reachable from any visited node
                    print("not visited", edge.get_node2().get_index())
                    if can_be_visited(edge.get_node2(), visited_node)==1:
                        print("can be visited")
                        visited_node.append(edge.get_node2())
                    else:
                        return -1
        return visited_node
    #print("not max so get next edge")
    for edge in node.get_edges():
        #print(edge.get_node1().get_index(), edge.get_node2().get_index())
        if (edge.get_node1().get_index() == node.get_index()):
            if(edge.get_node2() not in visited_node):
                #print("next node", edge.get_node2().get_index())
                returned=  recursive_search_on_graph(edge.get_node2(), max_node, visited_node)
                print("returned from", edge.get_node2().get_index(), returned)
                if returned != -1:
                    return returned
    visited_node.remove(node)
    return -1
def can_be_visited(node, visited_node):
    for e in node.get_edges():
        if(e.get_node2() in visited_node):
            visited_node.append(node)
            return 1
    return -1
```
